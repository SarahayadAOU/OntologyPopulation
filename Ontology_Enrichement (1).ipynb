```python
import csv

# Open the file and read from it within the 'with' block
with open("C:/Users/ayads/Downloads/Clean.csv", 'r') as csvfile:
    reader = csv.reader(csvfile)
    headers = next(reader)  # Move this inside the 'with' block
    print("Headers:", headers)  # Ensure this is also inside the 'with' block

    # Initialize the dictionary with headers
    columns = {}
    for header in headers:
        columns[header.strip()] = []  # Assuming you want to strip spaces from headers
        print ("here" , header)

    # Populate the dictionary with data from the CSV
    for row in reader:
        for header, value in zip(headers, row):
            stripped_header = header.strip()  # Again, assuming you want to strip spaces
            if stripped_header in columns:  # Check if the cleaned header is in columns
                columns[stripped_header].append(value)
            else:
                print(f"Missing header in columns: {stripped_header}")  # This should also be inside the 'with' block
```

Recursive Zero shot function with input id_p_comp

```python
import openai
import csv

openai.api_key = 'xxxxxxxxx'

def dynamic_Prompt_Zero_Shot_infile(csv_file_path, output_csv_path):
    # Initialize a set to store unique 'id_p_comp' values
    unique_id_p_comp = set()

    # Read 'id_p_comp' values from the input CSV and add them to the set
    with open(csv_file_path, mode='r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            unique_id_p_comp.add(row['id_p_comp'])

    # Open the output CSV file for writing
    with open(output_csv_path, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['id_p_comp', 'Response']) 

        # Iterate over each unique 'id_p_comp' value
        for id_p_comp in unique_id_p_comp:
            # Construct the prompt using the 'id_p_comp' value
            prompt = f"Provide other values for type_piece related to the product Glue for the building identified by {id_p_comp}."
            print(f"Prompt for {id_p_comp}: {prompt}")

            # Make the API call
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "system", "content": "You are a helpful assistant."},
                    {"role": "user", "content": prompt}
                ]
            )

            # Get the response content
            response_content = response.choices[0].message['content']
            print(f"Response for {id_p_comp}: {response_content}")

            # Write the 'id_p_comp' value and its corresponding response to the output CSV
            writer.writerow([id_p_comp, response_content])


dynamic_Prompt_Zero_Shot_infile('C:/Users/ayads/Downloads/Clean.csv', 'C:/Users/ayads/Downloads/Zero_Shot.csv')
```
